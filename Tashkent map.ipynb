```python
m = leafmap.Map(center=[37.7621, -122.4143], height=1200, zoom=12)

#csv_data = "C:/Users/Nikita.Dronov/OneDrive - Cushman & Wakefield/Desktop/Tashkent/dev/DATA TASHKENT GEO (1).csv"
#config_data = 'C:/Users/Nikita.Dronov/OneDrive - Cushman & Wakefield/Desktop/Tashkent/dev/out.json'
dataLink = "https://1drv.ms/x/s!AuM2Z_gMxN6KhuFxTGeBN3aVt_cE6Q?e=0vIZGG"
configLink = "out.json"
```

```python
def create_onedrive_directdownload (onedrive_link):
    data_bytes64 = base64.b64encode(bytes(onedrive_link, 'utf-8'))
    data_bytes64_String = data_bytes64.decode('utf-8').replace('/','_').replace('+','-').rstrip("=")
    resultUrl = f"https://api.onedrive.com/v1.0/shares/u!{data_bytes64_String}/root/content"
    return resultUrl

dirLink = create_onedrive_directdownload(dataLink)
```

```python
#let's download a df and split it into 3 separate dfs
df0 = pd.read_excel(dirLink, sheet_name="to_csv")
#df0 = pd.read_csv(csv_data, sep=';')

df_office = df0.loc[df0["Тип данных"]=="БЦ"]
df_mall = df0.loc[df0["Тип данных"]=="ТЦ"]
df_wi = df0.loc[df0["Тип данных"]=="Склад"]
df_bazaar = df0.loc[df0["Тип данных"]=="bazar"]
```
